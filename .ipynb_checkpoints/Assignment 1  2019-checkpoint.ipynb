**3. Obtain the 10 players with highest salary. (1 mark)**

**9. Get a subset of Golden State Warriors (GSW). Report the top 5 highest pay players of GSW. (1 mark)**

**GSW Subset

```python
gsw_subset = nba.loc[nba['team_name'] == "Golden State Warriors"]
gsw_subset
```

**Top 5 GSW Earners Report

```python
gsw_top5_earners = gsw_subset.nlargest(5, "salary")
gsw_top5_earners[['player', 'salary', 'team']]
```

**10. Get a subset of Cleveland Caveliers (CLE), the Runner-up team in the season. Who is the highest pay of CLE? Who is the lowest? How many times the highest-pay player than more than the lowest? (3 marks)**

**CLE Subset

```python
cle_subset = nba.loc[nba['team'] == "CLE"]
cle_subset
```

**Top CLE Earner

```python
cle_top_earner = cle_subset.nlargest(1, "salary")
cle_top_earner
```

**Least CLE Earner

```python
cle_least_earner = cle_subset.nsmallest(1, "salary")
cle_least_earner
```

**Difference between CLE Top and Least Earners

```python
top = cle_top_earner.salary.max()
least = cle_least_earner.salary.min()
top - least
```

**11. Combine the two subsets (GSW & CLE) created in questions 8 & 9. Find the top 4 pay players in the combined set. Compute the total salary of these 4 players. (3 marks)**

**Combined subset

```python
frames = [gsw_subset, cle_subset]
combined_subsets = pd.concat(frames)
combined_subsets
```

**The top 4 earners of the combined GSW and CLE subsets are thus

```python
combined_top4_earners = combined_subsets.nlargest(4, "salary")
combined_top4_earners
```

**Total salary for the top 4 paid players

```python
combined_top4_earners.salary.sum()
```
